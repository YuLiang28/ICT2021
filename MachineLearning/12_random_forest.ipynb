### 集成学习 ensemble learning
集成学习：
- 通过建立几个模型组合的解决单一预测问题
- 它的工作原理是生成多个分类器/模型，各自独立地学习和作出预测
- 这些预测最后合成组合预测，因此优于任何一个单分类作出的预测
- 三个臭皮匠顶个诸葛亮

### 随机森林
随机森林是一个包含多个决策树的分类器，并且其输出的类别是由多个树输出类别的众数所决定。  
例如，训练了5棵树，4课结果是 True，一颗结果是 False，那么最终结果也还是 True。  

#### 原理
- 两个随机：
  - 训练集随机 - N 个样本中随机有放回抽样 N 个
    - bootstrap 随机有放回抽样
  - 特征随机 - 从 M 个特征中随机抽取 m 个特征
    - M >> m 
    - M 远远大于 m
    - 起到降维的效果
### API
~~~python
sklearn.ensemble.RandomForestClassifier
~~~
- n_estimators: 树林里的树木数量，默认100
- criterion：默认是基尼系数(gini)，也可以选择信息增益的熵(entropy)
- max_depth: 树的深度
- max_features: 每个决策树的最大特征数量
  - int
  - float -> round(max_features * n_features)
  - auto -> max_features=sqrt(n_features)
  - sqrt -> max_features=sqrt(n_features) 与 auto 相同
  - log2 -> max_features=log2(n_features)
  - None -> max_features=n_features
- bootstrap: 是否在构建树时使用随机有放回抽样，默认 True
- min_samples_split: 叶节点最小划分样本数
- min_samples_leaf: 叶子节点的最小样本数
### 总结
- 准确性高
- 能有效运行在大数据集上，处理具有高维特征的输入样本，且不需要降维
- 能够评估各个特征在分类问题上的重要性
