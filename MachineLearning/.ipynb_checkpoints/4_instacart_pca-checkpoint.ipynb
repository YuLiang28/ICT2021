```python
# 获取数据
# 合并表
# 找出 user_id 和 aisle 之间的关系
# PCA 降维
```

```python
import pandas as pd
```

```python
# read data
order_products__prior = pd.read_csv(r"../datasets/instacart/order_products__prior.csv")
products = pd.read_csv(r"../datasets/instacart/products.csv")
orders = pd.read_csv(r"../datasets/instacart/orders.csv")
aisles = pd.read_csv(r"../datasets/instacart/aisles.csv")

```

```python
order_products__prior.head()
```

```python
products.head()
```

```python
orders.head()
```

```python
aisles.head()
```

```python
# concatenate table
# 将 aisle 和 order_id 合并到一张表中
aisles_products = pd.merge(aisles,products,on=["aisle_id","aisle_id"])
aisles_products.head()
```

```python
a_p_order_products__prior = pd.merge(aisles_products,order_products__prior,on=["product_id","product_id"])
a_p_order_products__prior.head()
```

```python
a_p_o_p_p_orders = pd.merge(a_p_order_products__prior,orders,on=["order_id","order_id"])
a_p_o_p_p_orders.head()
```

```python
mergeTable = a_p_o_p_p_orders
```

```python
mergeTable.head()
```

```python
# 交叉表
crossTable = pd.crosstab(mergeTable["user_id"],mergeTable["aisle"])
```

```python
crossTable
```

```python
# PCA 
from sklearn.decomposition import PCA
tf = PCA(n_components=0.95) # 保留 95% 的特征
pcaData = tf.fit_transform(crossTable)
pcaData
```

```python
pcaData.shape # 从 134 个特征降维到 44 个特征
```
