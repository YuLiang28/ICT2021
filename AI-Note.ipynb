```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
pd.set_option('display.max_rows', 500)
df = pd.read_csv('AI-Data.csv',encoding="ISO-8859-1")[:1000]

df
```

```python
# 清理空值
df = df.dropna(axis=0, how='any')

df["TotalStockPrice"] = df["Quantity"].mul(df["UnitPrice"])
df = df[df['InvoiceNo'].str.contains("C")==False]
df.head(5)
```

```python
tp = df.groupby(['InvoiceNo'])['TotalStockPrice'].agg('sum')

tp.name = "TotalPrice"
tp = tp.to_frame()
df = pd.merge(df, tp, on="InvoiceNo")

df
```

```python
# pd one hot encode
one_hot = pd.get_dummies(df['CustomerID'])
```

```python
df_lite = df[["UnitPrice","TotalPrice"]]

df_ohe = df_lite.join(one_hot)
df_ohe.head(5)
```

```python
X = df_ohe
X
```

```python
y = df[["Quantity"]]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
X_train
```

```python
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train,y_train)
```

```python
pred = reg.predict(X_test)
pred[:3]
```

```python
mean_absolute_error(y_test, pred)
```
